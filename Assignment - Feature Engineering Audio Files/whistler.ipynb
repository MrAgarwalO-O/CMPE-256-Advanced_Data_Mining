# **Importing required Libraries**

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

# **Reading the audio file**

```python
file_path = '/content/Parker Solar Probe - Whistler Mode Waves 2.mp3'
y, sr = librosa.load(file_path, sr=None)
```

# **Displaying the waveform**

```python
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

# **Display a spectrogram**

```python
plt.figure(figsize=(10, 4))
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.title('Spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.show()
```

# **Computing and displaying a mel-scaled spectrogram**

```python
plt.figure(figsize=(10, 4))
M = librosa.feature.melspectrogram(y=y, sr=sr)
M_db = librosa.amplitude_to_db(M, ref=np.max)
librosa.display.specshow(M_db, sr=sr, x_axis='time', y_axis='mel')
plt.title('Mel-scaled Spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.show()
```

# **Computing and displaying the Short-time Fourier transform**

```python
plt.figure(figsize=(10, 4))
D = librosa.stft(y)
librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), sr=sr, x_axis='time', y_axis='log')
plt.title('Short-time Fourier transform (STFT)')
plt.colorbar(format='%+2.0f dB')
plt.show()
```

# **Perform and display beat tracking**

```python
plt.figure(figsize=(10, 4))
tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
librosa.display.waveshow(y, sr=sr)
plt.vlines(librosa.frames_to_time(beats), -1, 1, color='r')
plt.title('Beat Tracking')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

# **Compute and display the constant-Q transform of the audio signal**

```python
plt.figure(figsize=(10, 4))
C = librosa.cqt(y, sr=sr)
librosa.display.specshow(librosa.amplitude_to_db(C, ref=np.max), sr=sr, x_axis='time', y_axis='cqt_note')
plt.title('Constant-Q transform (CQT)')
plt.colorbar(format='%+2.0f dB')
plt.show()
```
